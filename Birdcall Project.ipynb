```python
# Preparations for Deep Learning (from ChatGPT)
import cv2
import numpy as np
import os
import pandas as pd
from skimage.feature import local_binary_pattern
# Set path to your spectrogram images (e.g., 'train_images/' on Kaggle)
img_dir = "/Users/lorraineyang/Desktop/Summer/Birdcall Identification/Birdcall Dataset/example_test_image"
# List all image filenames
image_files = [f for f in os.listdir(img_dir) if f.endswith(".jpg") or f.endswith(".png")]
def extract_features(img_path):
    # Load image in grayscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    # Resize for consistency
    img = cv2.resize(img, (128, 128))
    # Pixel stats
    mean_val = np.mean(img)
    std_val = np.std(img)
    min_val = np.min(img)
    max_val = np.max(img)
    # Histogram of pixel intensities
    hist = cv2.calcHist([img], [0], None, [32], [0, 256])  # 32-bin histogram
    hist = hist.flatten() / np.sum(hist)  # normalize
    # Texture: Local Binary Pattern
    lbp = local_binary_pattern(img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 11), density=True)
    # Combine features into a single array
    features = np.concatenate([[mean_val, std_val, min_val, max_val], hist, lbp_hist])
    return features
X = []
y = []
for file in image_files:
    label = 1 if "corn_bunting" in file else 0  # adjust based on filename structure
    features = extract_features(os.path.join(img_dir, file))
    
    X.append(features)
    y.append(label)
# Turn into a DataFrame
df = pd.DataFrame(X)
df["label"] = y
df
```
